Substitua a palavra "NES" por "Nintendinho" no nome dos consoles e deixe todos os nomes em maiúsculos.
2.Filtre o nome dos consoles com data de release depois de 2010.
3.De um describe e info da base, substitua os missing values pela string "missing".
4.Filtre os consoles que foram descontinuados a menos de 2 anos da data de release.

```python
import pandas as pd

# Lê o arquivo Excel em um DataFrame
dataframe = pd.read_excel('best-selling game consoles.xlsx', sheet_name='consoles')

dataframe['Console Name'] = dataframe['Console Name'].str.replace('NES', 'Nintendinho').str.upper()

dataframe_filtrado = dataframe[dataframe['Released Year'] > 2010]


dataframe_filtrado.fillna("missing", inplace=True)

dataframe_filtrado

# Calcula a diferença entre as datas de descontinuação e lançamento
dataframe_filtrado['Diff Years Launch'] = dataframe_filtrado['Discontinuation Year'] - dataframe_filtrado['Released Year']

# Filtra os consoles descontinuados a menos de 2 anos do lançamento
consoles_descontinuados = dataframe_filtrado[dataframe_filtrado['Diff Years Launch'] <= 2]

consoles_descontinuados

# Exibe os consoles descontinuados a menos de 2 anos do lançamento
#print(consoles_descontinuados)


# Exibe informações sobre o DataFrame (describe e info)
# print(dataframe_filtrado.describe())
#  print(dataframe_filtrado.info())

# Exibe as primeiras linhas do DataFrame
#print(dataframe.head())
```
