```python
import pandas as pd
import seaborn as sns


# Criar um DataFrame de exemplo
data = {
    'Data': ['2023-08-01', '2023-08-01', '2023-08-02', '2023-08-02'],
    'Produto': ['A', 'B', 'A', 'B'],
    'Vendas': [100, 150, 200, 120]
}

df = pd.DataFrame(data)

# Pivotar a tabela
pivot_table = df.pivot(index='Data', columns='Produto', values='Vendas')

print(pivot_table)

sns.heatmap(pivot_table)
```
