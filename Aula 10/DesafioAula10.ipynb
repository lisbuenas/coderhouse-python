3. Sobre o dataframe anterior, adicione a coluna "Market Value" presente na tabela do arquivo "company.txt".

```python
import pandas as pd

# Lê o arquivo Excel em um DataFrame
dataframe = pd.read_excel('best-selling game consoles.xlsx', sheet_name='consoles')

# dataframe




# Gera a tabela de contagem por empresa usando agregação
tabela_contagem = dataframe.groupby('Company').size().reset_index(name='Units sold(million)')

tabela_ordenada = tabela_contagem.sort_values(by='Units sold(million)', ascending=False)

dataframe_company = pd.read_csv('company.txt', delimiter=',')

# dataframe_company

dataframe_completo = pd.merge(tabela_ordenada, dataframe_company, left_on='Company', right_on='Company')

dataframe_completo


# Exibe a tabela de contagem por empresa
# print(tabela_contagem)
```
